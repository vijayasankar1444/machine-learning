```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
dataframe = pd.read_csv("/content/1.01. Simple linear regression.csv")
print(dataframe.head())
X = dataframe['SAT'].values.reshape(-1, 1)
Y = dataframe['GPA']
plt.scatter(X, Y)
plt.show()
clf = LinearRegression()
clf.fit(X, Y)
print(f"R-squared score: {clf.score(X, Y)*100}")
prediction = clf.predict([[1664]])
print(f"Predicted GPA for SAT score 1664: {prediction}")
plt.scatter(X, Y)
plt.plot(X, clf.predict(X), color='blue')
plt.show()
```
