```python
from sklearn.svm import SVC
model=SVC()
model.fit(xtrain,ytrain)
```

```python
rbf_model = SVC(kernel='rbf')
```

```python
rbf_model.fit(xtrain, ytrain)
rbf_model.score(xtest,ytest)
```
