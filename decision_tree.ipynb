```python
from pandas import read_csv
import numpy as np
import sklearn
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import tree

dataframe = read_csv("/content/Heart.csv")
dataframe.head()
```

```python
dataframe.info()
```

```python
x=dataframe["Age"].values.reshape(-1,1)
y=dataframe["Gender"].values.ravel()
print(x,y)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf=DecisionTreeClassifier()
clf.fit(x,y)
print(clf.score(x,y))
```

```python
import numpy as np
x=np.array(x)
y=np.array(y)
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.10, random_state=4)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report
predictions = clf.predict(x_test)
print("precission values", predictions)
print("original value",y_test)
```

```python
print(classification_report(y_test, predictions))
```
